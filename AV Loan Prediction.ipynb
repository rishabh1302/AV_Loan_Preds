```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, f1_score
from sklearn.metrics import classification_report
```

```python
df_train = pd.read_csv('train_lp.csv',header=0,converters={'Self_Employed':str,'Property_Area':str,'Gender':str})
df_test = pd.read_csv('test_lp.csv',header=0,converters={'Self_Employed':str,'Property_Area':str,'Gender':str})
```

# Handling Missing Values, converting float to ints

```python
df_train.loc[df_train['Dependents'] == "3+", 'Dependents'] = 3
```

```python
df_train['Self_Employed'] = df_train['Self_Employed'].fillna("No")
df_train['Dependents'] = df_train['Dependents'].fillna(0)
df_train['LoanAmount'] = df_train['LoanAmount'].fillna(df_train['LoanAmount'].median())
df_train['Gender'] = df_train['Gender'].fillna("Male")
df_train['Married'] = df_train['Married'].fillna("No")
df_train['Credit_History'] = df_train['Credit_History'].fillna(1)
df_train['Loan_Amount_Term'] = df_train['Loan_Amount_Term'].fillna(360)
```

```python
df_train['CoapplicantIncome'] = df_train['CoapplicantIncome'].apply(np.int64) 
df_train['Loan_Amount_Term'] = df_train['Loan_Amount_Term'].apply(np.int64) 
df_train['Credit_History'] = df_train['Credit_History'].apply(np.int64)
df_train['Dependents'] = df_train['Dependents'].apply(np.int64)
df_train['LoanAmount'] = df_train['LoanAmount'].apply(np.int64) 


label_encoder = LabelEncoder()
df_train['Gender']= label_encoder.fit_transform(df_train['Gender'])
df_train['Property_Area']= label_encoder.fit_transform(df_train['Property_Area'])
df_train['Married']= label_encoder.fit_transform(df_train['Married'])
df_train['Dependents']= label_encoder.fit_transform(df_train['Dependents'])
df_train['Self_Employed']= label_encoder.fit_transform(df_train['Self_Employed'])

df_train.Education = df_train.Education.map({ 'Graduate' : 1, 'Not Graduate' : 0})
df_train.Loan_Status = df_train.Loan_Status.map({ 'Y' : 1, 'N' : 0})
```

```python
df_test.loc[df_test['Dependents'] == "3+", 'Dependents'] = 3
```

```python
df_test['Self_Employed'] = df_test['Self_Employed'].fillna('No')
df_test['Dependents'] = df_test['Dependents'].fillna(0)
df_test['LoanAmount'] = df_test['LoanAmount'].fillna(df_test['LoanAmount'].median())
df_test['Gender'] = df_test['Gender'].fillna("Male")
df_test['Married'] = df_test['Married'].fillna("No")
df_test['Credit_History'] = df_test['Credit_History'].fillna(1)
df_test['Loan_Amount_Term'] = df_test['Loan_Amount_Term'].fillna(360)
```

```python
df_test['CoapplicantIncome'] = df_test['CoapplicantIncome'].apply(np.int64) 
df_test['Loan_Amount_Term'] = df_test['Loan_Amount_Term'].apply(np.int64) 
df_test['Credit_History'] = df_test['Credit_History'].apply(np.int64)
df_test['Dependents'] = df_test['Dependents'].apply(np.int64)
df_test['LoanAmount'] = df_test['LoanAmount'].apply(np.int64) 

from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
df_test['Gender']= label_encoder.fit_transform(df_test['Gender'])
df_test['Property_Area']= label_encoder.fit_transform(df_test['Property_Area'])
df_test['Married']= label_encoder.fit_transform(df_test['Married'])
df_test['Dependents']= label_encoder.fit_transform(df_test['Dependents'])
df_test['Self_Employed']= label_encoder.fit_transform(df_test['Self_Employed'])

df_test.Education = df_test.Education.map({ 'Graduate' : 1, 'Not Graduate' : 0})
```

# Model Building

```python
train = df_train.drop('Loan_ID',axis=1)
test = df_test.drop('Loan_ID',axis=1)
```

```python
X = train.drop('Loan_Status',1)
y = train.Loan_Status
```

```python
from sklearn.model_selection import train_test_split

X_train,X_val,Y_train,Y_val = train_test_split(X,y,test_size=0.25)
```

```python
X_train
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.neighbors import KNeighborsClassifier 
from sklearn.ensemble import RandomForestClassifier
from lightgbm import LGBMClassifier
import xgboost as xgb
from xgboost.sklearn import XGBClassifier
```

```python
svc = SVC(kernel='rbf', gamma="scale")
knc = KNeighborsClassifier(n_neighbors=49)
mnb = MultinomialNB(alpha=0.2)
lrc = LogisticRegression(solver='liblinear', penalty='l2',C=10.)
rfc = RandomForestClassifier(n_estimators=500, random_state=111,n_jobs=-1,max_depth=10,max_features='sqrt')
lgbm = LGBMClassifier(n_estimators=200,learning_rate=0.3,max_depth=8,max_features="sqrt")
xgb = XGBClassifier(n_estimators=150,learning_rate=0.2)
```

```python
clfs = {'SVC' : svc,'KN' : knc, 'NB': mnb, 'LR': lrc, 'RF': rfc,'LGBM':lgbm,'XGB':xgb}
```

```python
def train(clf, features, targets):    
    clf.fit(features, targets)

def predict(clf, features):
    return (clf.predict(features))
```

```python
pred_scores_word_vectors = []
for k,v in clfs.items():
    train(v, X_train, Y_train)
    pred = predict(v, X_val)
    pred_scores_word_vectors.append((k, [accuracy_score(Y_val,pred)]))
```

```python
preds = {'Scores':pred_scores_word_vectors}
predictions = pd.DataFrame.from_dict(preds)
predictions
```

```python
pred=model.predict(test)
```
